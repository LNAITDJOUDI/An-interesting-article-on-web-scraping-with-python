```python
df['ville']=df['ville'].str.replace('  ',' ')
```

```python
# changé tous nc en nan
for col in list(df.columns):
    if col !='ville' and col!='lien':
        df.loc[df[col].astype(str).str.contains('nc',na=False),col]=np.nan
```

```python
# changé tous nc en nan
for col in list(df.columns):
    if col !='ville' and col!='lien':
          df[col]=df[col].astype(float)
```
